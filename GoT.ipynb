```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('script-bag-of-words.json')
```

```python
data.tail()
```

```python
data.iloc[-1]['text']
```

```python
dialouge = {}
for index, row in data.iterrows():
  for item in row['text']:
    if item['name'] in dialouge:
      #append
      dialouge[item['name']] = dialouge[item['name']] + item['text']
    else:
      #create character
      dialouge[item['name']] = item['text'] + " "
```

```python
new_data = pd.DataFrame()
new_data['character'] = dialouge.keys()
new_data['words'] = dialouge.values()
```

```python
new_data.iloc[:,0:3].head()
```

```python
new_data['num_words'] = new_data['words'].apply(lambda x:len(x.split()))
```

```python
new_data = new_data.sort_values('num_words', ascending=False)
```

```python
new_data = new_data.head(100)
```

```python
new_data.shape
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
embeddings = cv.fit_transform(new_data['words']).toarray()
```

```python
embeddings.shape
```

```python
embeddings = embeddings.astype('float64')
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, verbose=1, random_state=123)
z = tsne.fit_transform(embeddings)
```

```python
z.shape
```

```python
new_data['x'] = z.T[0]
new_data['y'] = z.T[1]
#new_data['z']=zT[2]
```
